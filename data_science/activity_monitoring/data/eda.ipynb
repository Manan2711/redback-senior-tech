```python
import pandas as pd

from data_augmentation import DataAugmentation
```

We will be using the following function to perform data augmentation and plot the results to visualise the difference between the original dataset and augmented data.

```python
def plot_augmentations(raw_data):
    """Plot original and augmented data for a given activity."""
    # Instantiate the DataAugmentation class with the merged data
    data_aug = DataAugmentation(raw_data)

    # List all augmentation methods
    augmentations = list(data_aug.augmentation_methods.keys())

    # Apply all augmentations and store in a dictionary
    augmented_datasets = {}
    for aug_name in augmentations:
        augmented_data = data_aug.augmentation_methods[aug_name]()
        augmented_datasets[aug_name] = augmented_data

    # Plot the first 300 samples for visualisation
    sample_length = 300

    # Prepare the data for plotting
    original_data_plot = raw_data[:sample_length, :]

    for aug_name, aug_data in augmented_datasets.items():
        # Ensure the augmented data is at least as long as sample_length
        aug_data_plot = aug_data[:sample_length, :]

        print(f"Visualising augmentation: {aug_name}")

        data_aug.plot_data(
            original_data_plot,
            aug_data_plot,
            aug_name,
            figsize=(18, 6))
```

We visualise the following data augmentation techniques:
* **Adding noise** – Used to introduce random noise to the data, helping the model become more robust to variations and preventing overfitting.
* **Scaling** – Alters the size of the input data, allowing the model to learn invariance and improving generalisation across different scales.
* **Time shift** – This involves moving the data along the time axis, which helps the model learn temporal relationships and become invariant to small shifts in time.
* **Flipping** – Flipping horizontally or vertically transforms the data, allowing the model to learn symmetrical features and increasing its ability to generalise across different orientations.
* **Random cropping** – This technique involves randomly selecting a subset of the data, which helps the model learn to focus on relevant features.
* **Magnitude warping** – This applies random distortions to the amplitude of the data, which can enhance the model's ability to recognize patterns despite changes in intensity.
* **Cutout/Masking** – This randomly removes sections of the data, forcing the model to learn to fill in gaps and enhancing its ability to generalize by preventing reliance on specific features

```python
# Load the demo activity folders
merged_data = pd.read_csv("demo_plot_data.csv").to_numpy()
plot_augmentations(merged_data)
```
